### Zadanie B

```python
def taskB(data):
    for p in range(len(data)):
        for i in range(len(data)-1):
            if abs(data[i]) > abs(data[i+1]):
                c = data[i]
                data[i] = data[i+1]
                data[i+1] = c
    
    print(data)
    return

taskB(test_vec)
```

### Zadanie C
Napisz i przetestuj funkcję wyznaczającą pierwiastki wielomianu kwadratowego opisanego wektorem wejściowym (uwaga - nie wolno wykorzystywać funkcji roots), oraz generującą jego wykres (w zakresie $<-4,4>$).

```python
import math

def taskC(polynom):
    delta = polynom[1]**2 - 4*polynom[2]*polynom[0]
    if delta > 0:
        x1 = (-1*polynom[1] + math.sqrt(delta))/2*polynom[2]
        x2 = (-1*polynom[1] - math.sqrt(delta))/2*polynom[2]
        print("Pierwiastki tego wielomianu to ",x1," i ",x2)
    elif delta==0:
        x1 = (-1*polynom[1] + math.sqrt(delta))/2*polynom[2]
        print("Pierwiastki tego wielomianu to ",x1)
    else:
        print("Ten wielomian nie ma pierwiastków rzeczywistych.")
    
    x = np.arange(-4, 4.1, 0.1)
    y = (test_poly[2]*x*x) + (test_poly[1]*x) + test_poly[0]
    
    plt.plot(x, y)
    plt.show()
    
    return

taskC(test_poly)
```

### Zadanie D
Napisz i przetestuj funkcję o argumentach będących współczynnikami wielomianu kwadratowego oraz zakresem i dokładnością punktów dziedziny, wykreślającą ten wielomian w podanym zakresie.

```python
def taskD(polynom, xrange, num_points):
    x = np.linspace(xrange[0], xrange[1], num_points)
    y = (polynom[2]*x*x) + (polynom[1]*x) + polynom[0]
    plt.plot(x, y)
    plt.show()
    return

taskD(test_poly, [-6, 4], 400)
```
